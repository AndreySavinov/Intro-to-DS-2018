# 4. Useful features

```python
from scipy.fftpack import fft
X_fft = fft(X)# I've tried to use descrete Fourier transform
```

```python
X_fft_plus = X_fft[y_plus]
X_fft_minus = X_fft[y_minus]
```

## 4.1 Average absolute value of Fourier coefficients and Standart Deviation. If Absolute Value is more than 8.5 than score decreses by 1, in other case increases by 1.  If Standart Deviation is less than 20.66 than score decreses by 1, in other case increases by 1. If at the end score is positive, class is +1. Score is a class defining function. 

```python
plt.figure(figsize=(15,5), facecolor='pink', frameon=True)
plt.subplot(1, 2, 1)
sns.distplot(np.mean(abs(X_fft_minus), axis = 1), hist = False);
sns.distplot(np.mean(abs(X_fft_plus), axis = 1), hist = False);
plt.xlabel('Mean Value of FC');

plt.subplot(1, 2, 2)
sns.distplot(np.std(abs(X_fft_minus), axis = 1), hist = False)
sns.distplot(np.std(abs(X_fft_plus), axis = 1), hist = False)
plt.xlabel('Standart Deviation of FC');# FC = Fourier Coefficients
```

## 4.2 According to the graphic of mean absolute value of Fourier coefficients of -1 and +1 classes, coefficients of -1 and +1 classes differ from each other in a range from 25 harmonic and 50 harmonic and near 15 harmonic. I've tried to use the absolute meaning of some coefficients in order to make additional rules.

```python
plt.plot(np.mean(abs(X_fft_minus[:, 0:81]), axis = 0))
plt.plot(np.mean(abs(X_fft_plus[:, 0:81]), axis = 0))
plt.xlabel('Harmonic Number');
plt.ylabel('Mean Value CF for class')
plt.xticks(range(0,81, 10));
```

```python
plt.figure(figsize=(20,5), facecolor='pink', frameon=True)# Distribution of absolute values of some harmonics in classes
plt.subplot(1, 3, 1)
sns.distplot(abs(X_fft_minus[:, 14]), hist = False);
sns.distplot(abs(X_fft_plus[:, 14]), hist = False);
plt.xlabel('14 harmonic');
plt.xticks(range(0, 141, 20));
plt.xlim([-10, 150]);

plt.subplot(1, 3, 2)
sns.distplot(abs(X_fft_minus[:, 28]), hist = False);
sns.distplot(abs(X_fft_plus[:, 28]), hist = False);
plt.xlabel('28 harmonic');
plt.xticks(range(0, 141, 20));
plt.xlim([-10, 150]);

plt.subplot(1, 3, 3)
sns.distplot(abs(X_fft_minus[:, 38]), hist = False);
sns.distplot(abs(X_fft_plus[:, 38]), hist = False);
plt.xlabel('38 harmonic');
plt.xticks(range(0, 101, 20));
plt.xlim([-10, 100]);
```

### According to these graphs I've chosen the concrete vaules of absolute Fourier coefficients for each harmonic. Concrete values are accounted for in Score function. 

# 5. Desicion rule and accuracy calculation

### Score is a fucntion that increses or decreases for 1 according to the values of features

```python
def score_(X):
    sc = 0
    X_fft = fft(X)
    
    if (np.mean(abs(X_fft)) > 8.5):
        sc = sc - 1
    else:
        sc = sc + 1
        
    if (np.std(abs(X_fft)) < 20.66):
        sc = sc - 1
    else:
        sc = sc + 1
    
    if (abs(X_fft[28]) < 45):
        sc = sc - 1
    else:
        sc = sc + 1
    
    if (abs(X_fft[14]) > 40):
        sc = sc - 1
    else:
        sc = sc + 1
    
    if (abs(X_fft[38]) > 15):
        sc = sc - 1
    else:
        sc = sc + 1
       

    if (sc < 0):
        return -1
    else:
        return 1
```

```python
y_test = np.array(test_data['Target'])
X_test = np.array(test_data.drop('Target', axis=1))
```

```python
y_pred = []
for i in range(3601):
    y_pred.append(score_(X_test[i, :]))
y_pred = np.array(y_pred)
```

```python
print('Accuracy of this model is', accuracy_score(y_test, y_pred))
```
